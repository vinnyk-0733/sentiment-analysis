```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="t5-small")
```

```python
feedback = """The product arrived late and packaging was damaged. 
Although the product itself works fine, the delay and customer service experience were disappointing."""
```

```python
short_summary = summarizer(feedback, max_length=30, min_length=10, do_sample=False)
```

```python
detailed_summary = summarizer(feedback, max_length=60, min_length=20, do_sample=False)
```

```python
print("Short Summary:", short_summary[0]['summary_text'])
print("Short Summary:", detailed_summary[0]['summary_text'])
```
